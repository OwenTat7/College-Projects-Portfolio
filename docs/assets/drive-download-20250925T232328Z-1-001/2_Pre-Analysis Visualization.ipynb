```python
import pandas as pd
df = pd.read_csv("/Users/NathanSwan 1/Downloads/heart_cleaned_final2.csv", nrows=50000)
df.info()
df.head()
```

1. Distribution of the Dependent Variable. 0 meaning they do not have heart disease and 1 meaning they do have heart disease

```python
import seaborn as sns
import matplotlib.pyplot as plt

hd_counts = df['HeartDisease'].value_counts().sort_index()
plt.figure(figsize=(6,4))
plt.bar(hd_counts.index, hd_counts.values)
plt.xticks([0,1])
plt.xlabel('Heart Disease Presence')
plt.ylabel('Count')
plt.title('Distribution of Heart Disease')
plt.show()
```

2. Smoking Status. 0 meaning that the patient does not smoke and 1 meaning that they do smoke.

```python
smoke_counts = df['Smoking'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(smoke_counts.index, smoke_counts.values)
plt.xticks([0,1])
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.title('Distribution of Smoking Status')
plt.show()
```

3. BMI Distribution By Heart Disease Status

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="BMI", hue="HeartDisease", kde=True, bins=40, palette="coolwarm", element="step")
plt.title("BMI Distribution by Heart Disease Status")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

4. Alcohol Drinking Status. 0 means they do not drink and 1 means they do drink.

```python
alc_counts = df['AlcoholDrinking'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(alc_counts.index, alc_counts.values)
plt.xticks([0,1])
plt.xlabel('Alcohol Drinking Status')
plt.ylabel('Count')
plt.title('Distribution of Alcohol Drinking')
plt.show()
```
